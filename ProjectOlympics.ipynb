# Plot the height and weight of the participants in olympic games

```python
import matplotlib.pyplot as plt
df[["Height","Weight"]]=df[["Height","Weight"]].fillna(0).astype(int)
plt.figure(figsize=(10,10))
box_data=df[["Height","Weight"]]
labels=["Height of the participants","Weight of the participants"]
color=['r','c']
box=plt.boxplot(box_data,sym='p',labels=labels,patch_artist=True)
for patch,color in zip(box['boxes'],color):
    patch.set_facecolor(color)
plt.title("Height and Weight plot")
plt.show()
```

# Plot the graph for the last 10 years gold winning items for United States

```python
import numpy as np
df3=np.where((df["Year"]>=2000) & (df["Medal"]=="Gold") & (df["Team"]=="United States"))
df31=df.loc[df3]
df31["Sport"].value_counts()
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,18))
sns.countplot(y="Sport",data=df31)
plt.title("Last 10 years Gold winning items of United States")
```

# Plot the graph for sports where participants above 60 age won medal in olympics

```python
plt.figure(figsize=(10,10))
values1=df[df["Age"]>=60]["Sport"].value_counts()
plt.plot(values1,'go--', linewidth=2, markersize=8)
plt.xlabel("Sports item name")
plt.ylabel("Number of people won medal",loc="center")
plt.title("The Sports items elder participants won medal in olypics",loc="center")
plt.grid()
plt.show()
```

# Find the medals won by each team on Men's Football that was held in London and the age of participants between 20 and 35.

```python
df8=np.where((df["Age"]>=20) & (df["Age"]<=35) & (df["Sex"]=="M") & (df["Event"]=="Football Men's Football") & (df["City"]=="London"))
df81=df.loc[df8]
df81[df81["Medal"]=="Gold"]["Team"]
for i in df81["Medal"].unique():
    print(f"Number of {i} medal winners of Men's Football held in London and their age between (20-35) is :",len(df81))
    print(f"And number of {i} medals won by each team is ")
    print(df81["Team"].value_counts())
    print("")
    print("")
```

# Plot pie chart for all medal winners for each male and female participants

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
values=df[df["Medal"]=="Gold"]["Sex"].value_counts()
color=("r","green")
label=("Male","Female")
plt.pie(values,colors=color,labels=label,autopct="%0.2f%%")
plt.title("Male and Female Gold Medal winners Pie Chart")

plt.subplot(2,2,2)
values=df[df["Medal"]=="Silver"]["Sex"].value_counts()
color=("r","green")
label=("Male","Female")
plt.pie(values,colors=color,labels=label,autopct="%0.2f%%")
plt.title("Male and Female Silver Medal winners Pie Chart")

plt.subplot(2,2,3)
values=df[df["Medal"]=="Bronze"]["Sex"].value_counts()
color=("r","green")
label=("Male","Female")
plt.pie(values,colors=color,labels=label,autopct="%0.2f%%")
plt.title("Male and Female Bronze Medal winners Pie Chart")

plt.subplot(2,2,4)
values=df["Sex"].value_counts()
color=("r","green")
label=("Male","Female")
plt.pie(values,colors=color,labels=label,autopct="%0.2f%%")
plt.title("Pie chart for Male and Female participants in Olympics")

plt.suptitle("Sub Plots")
plt.show()
```
